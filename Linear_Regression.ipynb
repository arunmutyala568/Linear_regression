```python
import pandas as pd
import os 
import numpy as np
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

data=pd.read_csv('D:\Salary_Data.csv',sep=",",columns=['arun','arun'])


x=data[["YearsExperience"]]

y=data.Salary

lm = LinearRegression()
lm.fit(x, y)



xmean=np.mean(data.YearsExperience)

ymean=np.mean(data.Salary)

xcofficient=[]
denom=[]

pred=[]

ycoff=[]

ycoff2=[]

for i in range(len(data)):
    
    xcofficient.append((data['YearsExperience'][i]-xmean)*(data['Salary'][i]-ymean))
    denom.append((data['YearsExperience'][i]-xmean)**2)
    
num=np.array(xcofficient).sum()

den=np.array(denom).sum()

coef=num/den

intercept=(ymean-(num/den)*xmean)


print(coef)


print(intercept)

for i in range(len(x)):
    
    
    pred.append(coef*(data['YearsExperience'][i])+intercept)
    
    
print(pred)


def calaculater2():
    
    
    for i in pred:
        
        
        ycoff.append((i-ymean)**2) 
    ynum=np.array(ycoff).sum()
    for i in range(len(data)):
        
        ycoff2.append((data["Salary"][i]-ymean)**2)
        
    yden=np.array(ycoff2).sum()
   
    print(ynum/yden)
    
    
print(" ")
calaculater2()

print(lm.score(x,y))

predex=lm.predict(data[['YearsExperience']])
np.set_printoptions(suppress=True)
print(predex)
plt.scatter(data.YearsExperience,data.Salary)
plt.plot(data.YearsExperience,predex,color="red")
plt.legend()
print(coef*(6.0)+intercept)    
print(data)

print(lm.intercept_)

lm = smf.ols(formula='Salary ~ YearsExperience', data=data).fit()
print(lm.conf_int())

data.iloc[1,2]
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt

boston = load_boston()
bos = pd.DataFrame(boston.data,columns=boston.feature_names)

bos.isnull().sum()
bos['MEDV'] = boston.target

lm=LinearRegression()

x=bos[['CRIM']]
y=bos.MEDV
lm.fit(x,y)

predication=lm.predict(bos[['MEDV']])
print(lm.score(x,y))
plt.scatter(bos.CRIM,bos.MEDV ,color="red")
plt.legend()
plt.show()
```
